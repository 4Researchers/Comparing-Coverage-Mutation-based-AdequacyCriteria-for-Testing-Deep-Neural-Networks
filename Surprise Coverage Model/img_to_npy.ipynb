```python
import os
#from PIL import image
import numpy as np
import cv2
from keras.preprocessing.image import load_img,img_to_array
```

```python
for j in range(0,10):
    for i in range(0,101):
        graph_path = 'Documents/icse2019/build-in-resource/dataset/mnist/adversarial/jsma2'+'/r'+str(j)+'/a'+str(i)
        graph_list = os.listdir(graph_path)
        filenames = []
        graph_array = []
        for p in graph_list:
            filenames = os.path.join(graph_path,p)
            img = load_img(filenames,color_mode="grayscale",target_size=(28,28))
            #img = load_img(filenames,target_size=(32,32))
            img_arr = img_to_array(img)
            graph_array.append(img_arr)
        graph_array=np.array(graph_array)
        print(len(graph_array))
        path = 'notebooks/sadl-master/npydata/mnist/jsma2/r{0}'.format(j)
        if not os.path.isdir(path):
            os.makedirs(path)
        np.save(path+'/z'+str(i),graph_array)
```
